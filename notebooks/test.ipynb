```python
# Import own functions
from data.phantoms import generate_shepp_logan
from data.sinogram import forward_project

# Import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = generate_shepp_logan(size = 100)

# Create a sinogram from the Shepp-Logan phantom
sinogram = forward_project(image = x, angles = np.linspace(0, 180, 180, endpoint=False), noise = 0.01)

# Create figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(9, 9))

# Plot original image
axs[0,0].imshow(x, cmap='gray')
axs[0,0].set_title('Original Image')
axs[0,0].axis('off')

# Plot sinograms with different noise levels
for i, noise in enumerate([0.01, 0.05, 6]):
    sino = forward_project(image=x, angles=np.linspace(0, 180, 180, endpoint=False), noise=noise)
    row = (i+1) // 2 
    col = (i + 1) % 2 
    print(col)
# Skip (0,0) as it's already used for original image
    axs[row, col].imshow(sino, cmap='gray', aspect='auto')
    axs[row, col].set_title(f'Sinogram (noise={noise})')
    axs[row, col].axis('off')

plt.tight_layout()
plt.show()
```
